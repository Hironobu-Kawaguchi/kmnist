# Kuzushiji-MNIST (KMNIST) EDA

```python
import os
import numpy as np

path = '../input'
train_images = np.load(os.path.join(path, 'kmnist-train-imgs.npz'))['arr_0']
train_labels = np.load(os.path.join(path, 'kmnist-train-labels.npz'))['arr_0']
test_images = np.load(os.path.join(path, 'kmnist-test-imgs.npz'))['arr_0']
test_labels = np.load(os.path.join(path, 'kmnist-test-labels.npz'))['arr_0']
```

```python
import pandas as pd
kmnist_classmap = pd.read_csv(os.path.join(path, 'kmnist_classmap.csv'))
names = dict(zip(kmnist_classmap.index, kmnist_classmap['char']))
```

```python
names
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['Hiragino Maru Gothic Pro', 'Yu Gothic', 'Meirio', 'Takao', 'IPAexGothic', 'IPAPGothic', 'VL PGothic', 'Noto Sans CJK JP']
```

## 画像確認

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(names[train_labels[i]])
```

## 平均画像

```python
plt.figure(figsize=(12, 9))
for label in range(10):
    plt.subplot(3, 4, label + 1)
    indices = np.where(train_labels == label)
    mean_image = train_images[indices].mean(axis=0)
    plt.imshow(mean_image, cmap='gray')
    plt.title(names[label])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
